
Normalizzare significa trasformare i dati in modo che abbiano una scala comune o uno standard. Questo processo è spesso utilizzato quando si lavora con dati che variano notevolmente in scala o che hanno distribuzioni diverse. La normalizzazione aiuta a rendere i dati comparabili e facilita l'analisi.

Nel contesto della normalizzazione delle altezze nel codice che hai fornito:

1. **Sottrazione della media (`np.mean(altezze)`)**: Questo passaggio fa sì che la media delle altezze normalizzate sia zero. Sottrarre la media ai dati centra i dati attorno allo zero.

2. **Divisione per la deviazione standard (`np.std(altezze)`)**: Questo passaggio scala i dati dividendoli per la deviazione standard. Ciò standardizza i dati in modo che abbiano una dispersione unitaria. Questo è particolarmente utile quando si lavora con dati che possono avere variazioni significative nella loro scala.

La formula generale per normalizzare un insieme di dati è:

X_normalizzato =( X -media(X))\deviazione standard(X)

Dove:
- X_normalizzato è l'insieme normalizzato di dati.
- X è l'insieme originale di dati.
- media(X) è la media dell'insieme \(X\).
- deviazione standard(X) è la deviazione standard dell'insieme \(X\).

La normalizzazione è utilizzata in vari contesti, inclusi quelli statistici, di apprendimento automatico e di analisi dei dati, per garantire che le differenze nelle scale dei dati non influenzino indebitamente i risultati delle analisi o degli algoritmi.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
altezze=np.random.normal(176,5,30)
altezze
```

```python
np.mean(altezze)
```

```python
np.std(altezze)
```

```python
temp=altezze-np.mean(altezze)
temp
```

```python
np.std(temp)
```

```python
altezze_normalizzate=(altezze-np.mean(altezze))/np.std(altezze)
np.mean(altezze_normalizzate)
```

altezze_normalizzate = (altezze - np.mean(altezze)) / np.std(altezze): Questa linea di codice calcola le altezze normalizzate. Per farlo, sottrae la media delle altezze (np.mean(altezze)) da ciascuna altezza nell'array altezze e divide il risultato per la deviazione standard delle altezze (np.std(altezze)). La normalizzazione aiuta a portare i dati su una scala comune, rendendo più facile confrontare e interpretare i valori.

np.mean(altezze_normalizzate): Questa linea di codice calcola la media delle altezze normalizzate. La media normalizzata dovrebbe essere molto vicina a zero, poiché la normalizzazione ha l'obiettivo di centrare i dati attorno allo zero. Tuttavia, potrebbe esserci una leggera discrepanza dovuta alla precisione numerica del calcolo.

```python
np.std(altezze_normalizzate)
```

```python
fig,axs=plt.subplots(nrows=1,ncols=2,figsize=(8,4))
axs[0].hist(altezze,bins=10)
axs[1].hist(altezze_normalizzate, bins=10)
plt.show()
```

## ANALISI GRAFICA DATASET PIGUINI
l'analisi grafica di un dataframe è un grafico che contiene gli n*n grafici delle variabili nuemeriche del dataset

ci sono tutte le combinazioni delle colonne e n istogrammi quante sono le variabili

```python
pinguini=pd.read_csv('penguins_size.csv')
pinguini.head()
```

```python
marziani=pd.read_csv('marziani.csv')
marziani.head()
```

```python
def analisi_grafica(target,dati,features=None,dropna=True,alpha=1):
    
    colonne=np.array(dati.columns)
    colonne_num=[]
    
    if features==None:
        for colonna in colonne:

            if (str(type(dati[colonna][1]))[8:19] == 'numpy.float') or (str(type(dati[colonna][1]))[8:17] == 'numpy.int'):
                
                colonne_num.append(colonna)
        
    else:
        colonne_num=features
        
    colonne_num=np.array(colonne_num).flatten()
    dataframe=dati.copy()
    if dropna:
        dataframe=dataframe.dropna(axis="index",subset=colonne_num)

    col_specie={}
    for specie in dataframe[target]:
        col_specie[specie]=np.random.uniform(0,1,3)
    colori=[]
    for specie in dataframe[target]:
        colori.append(col_specie[specie])
    colori=np.array(colori)

    n=np.size(colonne_num)
    fig, axs=plt.subplots(ncols=n, nrows=n, figsize=(15,17))
    for riga in range(n):
        for colonna in range(n):
            if riga==colonna:
                for specie in col_specie.keys():
                    axs[riga, colonna].hist(dataframe[dataframe[target]==specie][colonne_num[riga]].values, color=col_specie[specie], label=specie,alpha=alpha)
                    axs[riga, colonna].set_xlabel(colonne_num[riga])
                    axs[riga, colonna].set_ylabel("Frequenza")
                    axs[riga, colonna].legend()
            else:
                for specie in col_specie.keys():
                    axs[riga, colonna].scatter(dataframe[dataframe[target]==specie][colonne_num[riga]].values, dataframe[dataframe[target]==specie][colonne_num[colonna]].values, color=col_specie[specie], s=10, label=specie)
                    axs[riga, colonna].set_xlabel(colonne_num[riga])
                    axs[riga, colonna].set_ylabel(colonne_num[colonna])
                    axs[riga, colonna].legend()

    plt.show()
```

```python
analisi_grafica("species",pinguini)
```

```python
analisi_grafica("species",pinguini,features=["culmen_length_mm","culmen_depth_mm","flipper_length_mm","body_mass_g"],alpha=0.5)
```

```python
analisi_grafica("specie",marziani)
```
