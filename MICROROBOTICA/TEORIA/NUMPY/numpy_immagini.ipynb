```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
altezza=100
larghezza=100
immagine=np.zeros((altezza,larghezza),dtype=np.int16)
#se voglio dare un colore diverso alla prima riga
immagine[0:5]=64
#cambio colore all'ultima riga
immagine[-5:]=128
#cambio colore alla prima e ultima colonna
immagine[:,0:5]=192
immagine[:,-5:]=256
```

```python
plt.imshow(immagine)
```

```python
##creo un quadrato al centro di lato venti
lato=20
immagine[altezza//2-lato//2:altezza//2+lato//2:,larghezza//2-lato//2:larghezza//2+lato//2]=512
```

```python
plt.imshow(immagine)
```

```python
altezza=100
larghezza=100
immagine=np.zeros((altezza,larghezza),dtype=np.int16)
```

```python
plt.imshow(immagine)
```

```python
n_pixel=100
x=np.random.randint(0,altezza,n_pixel)
y=np.random.randint(0,altezza,n_pixel)
immagine[x,y]=256
```

```python
plt.imshow(immagine)
```
