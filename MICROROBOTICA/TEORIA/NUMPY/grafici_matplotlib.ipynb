##### UN GRAFICO è COPOSTO
- titolo

- ogni asse ha un'etichetta dove scriviamo cosa l'asse rappresenta
- etichette sulle tacche: indicano i valori

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#grafico di una funzione
#si può fare il disegno di un grafico calcolando la funzione in determinati punti
#per creare un array di punti in un intervallo
x=np.linspace(0.,4*np.pi,100)
#se io campiono poche volte ho pochi valori di x, poi li unisce con segmenti, più ne ho + è preciso

y1=np.sin(x)
y2=np.cos(x)
```

```python
fig, ax=plt.subplots(figsize=(6,3))#usa la funzione subplots-> prende vari parametri, 
#figsize determina la dimensione della figura in pollici-> 1 poillice=2,54 cm
#questa chiamata mi restituisce la figura, fig e gli assi

#posso inserire un titolo
ax.set_title("un grafico a linea",fontsize=10)

#settiamo le etichette agli assi

ax.set_xlabel("x")
ax.set_ylabel("y")

ax.plot(x,y1, "r--",label="sin(x)", linewidth=3)
#tutte le chiamate per disegnare il grafico o modificare lo stile sono chiamate sull'oggetto ax
#il metodo plot crea un grafico a linea
#subio dopo aver specificato i dati posso speificare il formato, es "r" viene rossa, - linea continua, -- tratteggio, -.
ax.plot(x,y2,"k+", label="cos(x)", linewidth=3)
#per aumentare lo spessore delle curve-> linewidth
#faccio tanti ax.plot per ogni linea che voglio rappresentare

#per creare una legenda: -> devo inserire le label sopra per usare la lagenda
ax.legend()
ax.grid()#creo la griglia
plt.show()
```

```python
## esercizio
#consumi l/100km
#emissioni gr/100 km
#grafico a dispersione che leghi le emissioni ai consumi
#coloriamo i punti in base alla marca dell'auto

'''
x benzina normale
z super
d diesel
e etanolo
m metano
'''


file=open("./Fuel_Consumption_2000-2022.csv","r",encoding="utf-8")
righe=file.readlines()
file.close()
consumi=[]
emissioni=[]
produttori=[]
combustibile=[]

righe=righe[1:]
righe=[elemento.replace('\n', '') for elemento in righe]
for riga in righe:
    campi=riga.split(",")
    consumi.append(float(campi[8]))
    emissioni.append(float(campi[-1]))
    produttori.append(campi[1])
    combustibile.append(campi[7])
    
consumi=np.array(consumi)
emissioni=np.array(emissioni)
produttori=np.array(produttori)
combustibile=np.array(combustibile)
#np.max(emissioni)
produttori
```

```python
colori_prod={}
for produttore in produttori:
    if produttore not in colori_prod.keys():
        colori_prod[produttore]=(np.random.randint(0,256),np.random.randint(0,256),np.random.randint(0,256))

colori_carb={}
for c in combustibile:
    if c not in colori_carb.keys():
        colori_carb[c]=(np.random.randint(0,256),np.random.randint(0,256),np.random.randint(0,256))
        
colori= np.zeros(np.size(produttori), dtype=[('x', int), ('y', int), ('z', int)])

for i, produttore in enumerate(produttori):
    colori[i] = colori_prod[produttore]
    
colori_c= np.zeros(np.size(produttori), dtype=[('x', int), ('y', int), ('z', int)])

for i, c in enumerate(combustibile):
    colori_c[i] = colori_carb[c]
    
colori_c
```

```python
fig, ax=plt.subplots(figsize=(5,5))
colori = [(x/255, y/255, z/255) for (x, y, z) in colori]
ax.scatter(emissioni, consumi,s=5,c=colori)#,c=colori,s=100,marker="*",zorder=10
# lo uso per fare i grafici a disperisone
#uso c per i colori
#uso s per la dimensione
#zorder=10 per disegnare la grilia sotto le stelle, è lordine con cui disegna le figure, + è grande + lo disegna dopo

ax.grid()
ax.set_xlabel("emissioni gr/100 km")
ax.set_ylabel("consumi l/100km")


plt.show()
```

```python
#estrarre le 5 marche con più auto, funzioni che conosco
n_marche=5

num_auto_produttore={}
for produttore in produttori:
    if produttore not in num_auto_produttore.keys():
        num_auto_produttore[produttore]=np.size(produttori[produttori==produttore])

num_auto = np.array(sorted(num_auto_produttore.values(), reverse=True))

prod_max=[]
for i in range (n_marche):
    for chiave, valore in num_auto_produttore.items():
        if num_auto[i] == valore and chiave not in prod_max:
            prod_max.append(chiave) 
prod_max=np.array(prod_max)

comb_max=[]
for i in combustibile:
    if i not in comb_max:
        comb_max.append(i)
comb_max=np.array(comb_max)    

consumi2 = consumi[np.array([p in prod_max for p in produttori])]
emissioni2 = emissioni[np.array([p in prod_max for p in produttori])]
prod = produttori[np.array([p in prod_max for p in produttori])]

colori2= np.zeros(np.size(produttori), dtype=[('x', int), ('y', int), ('z', int)])
colori2 = colori[np.array([p in prod_max for p in produttori])]

colori_carburante=colori_c[np.array([p in prod_max for p in produttori])]
combustibile2=combustibile[np.array([p in prod_max for p in produttori])]

for i, produttore in enumerate(prod):
    colori2[i] = colori_prod[produttore]
    #print(colori2[i] )
#np.size(emissioni2)
```
