```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from datetime import datetime
from sklearn.preprocessing import StandardScaler
```

This dataset has been created from a query done on an website specialized in used cars
Description of colums:
- model: Fiat 500 comes in several 'flavours' :'pop', 'lounge', 'sport', 'star'
- engine_power: number of horse power
- transmission: manual or automatic shift
- age_in_days: age of the car in number of days (from the time the dataset has been created)
- km: kilometers of the car
- previous_owners: number of previous owners
- lat: latitude of the seller (the price of cars in Italy varies from North to South of the country)
- lon: longitude of the seller (the price of cars in Italy varies from North to South of the country)
- price: selling price (the target)

```python
auto=pd.read_csv("Used_fiat_500_in_Italy_dataset.csv")
auto
```

```python
auto.info()
```

```python
ar_mod=auto["model"].copy()
ar_mod
```

```python
modelli={}
for i, mod in enumerate(np.unique(auto["model"])):
    modelli[mod]=i
    ar_mod[ar_mod==mod]=i
    
modelli
ar_mod
```

```python
auto["modello_num"]=ar_mod
```

```python
auto
```

```python
ar_trasm=auto["transmission"].copy()
ar_trasm
```

```python
trasm={}
for i, tr in enumerate(np.unique(auto["transmission"])):
    trasm[tr]=i
    ar_trasm[ar_trasm==tr]=i
auto["trasm_num"]=ar_trasm
```

```python
auto
```

```python
X=auto[["engine_power","age_in_days","km","previous_owners","lat","lon","modello_num","trasm_num"]].values
y=auto["price"].values
```

```python
regressione1=LinearRegression()
regressione1.fit(X,y)
print(f"R^2 vale {regressione1.score(X,y):.2f}")
```

```python
X=auto[["engine_power","km","modello_num","trasm_num"]].values
y=auto["price"].values
regressione1=LinearRegression()
regressione1.fit(X,y)
print(f"R^2 vale {regressione1.score(X,y):.2f}")
```

```python
regressione1.predict([[69,4474,800000,2,45.071079,7.46403,1,1]])
```
