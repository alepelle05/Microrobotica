### regressioni lineari (cap. 10)
caso più semplice di algoritmo che è in grado di imparare e dare in output dei numeri, sviluppata in parte da GAUS stesso, ai tempi le regressioni lineari venivano fatte completamente a mano

le regressioni ci permettono di creare algoritmi che prevedono dei valori, trend

sugli algoritmi di machine learning le regressioni sono molto particolari

esempio:
abbiamo un dataset di training fatto da 2 colonne: x,y
y:target
x:feature

possiamo creare un grafico a dispersione dove rappresentiamo la y rispetto alla x

-> la regressione lineare prova a trovare la retta che meglio approssima questi punti nel piano x, y

se ho una retta e un punto sulla x so calcolare y, quindi la regressione mi sa calcolare cosa succede in un punto che non ho mai visto, lui in base alla retta sa gestire tutti i nuovi casi, testo il codice

eq retta -> y=ax+b -> bisogna trovare i migliori a e b per approssimare al meglio i dati

- una retta approssima al meglio i dati quanto più i residui sono vicini allo 0
per vedere se una retta approssima i dati si calcola il residuo di ogni punto, ossia la distanza lungo l'asse y tra il punto alla retta

#### algoritmo dei minimi quadrati (ols: ordinary least square)
per ogni punto calcolo la distanza dalla retta, gli elevo al quadrato e faccio la somma dei quadrati dei residui per far si che sia la minima possibile, è la più piccola possibile solo nel caso della retta che approssima meglio i dati

inventato da gaus, si riesce a risolvere con carta e penna, grazie alla regressione si è scoperto plutone
