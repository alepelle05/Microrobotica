## reti neurali

nelle reti neurali l'idea di base è il neurone del nostro cervello.

la parte esterna del cervello è rivestita da un rivestimento di circa 1 cm, è la neocorteccia che è la parte che svoge le funzioni della neointelligenza, li si trovano i neuroni che sono cellule molto particolari.

invecchiando queste cellule muoiono e non si riproducono più, sono fatte da una parte chiamata soma (dove si trova il nucleo) e una membrana cellulare detta dendridi con cui si collega con delle sinapsi ad altri neuroni.

i neuroni sono collegati tramite assoni a tutto ciò che è un input o un output per il nostro cerello, i segnali all'interno dell'assone sono elettrici condotti da soluzioni elettrolitiche

esperimento: mettendo un casco su una persona che rilevasse le onde celebrali si è riuscito a creare un algoritmo che mediante le onde riuscirre a capire cosa quella persona vedeva.

paloencefalo: meccanismi elementari, esempio la fame

il soma, il nucleo del neurone può essere schematizzato da una formula matenatica, i dentriti sono dei input e gli assoni dei output, questo compone il *neurone artificiale*


ogni neurone è collegato contemporaneamente a un numero che va da 1000 a 10000 sinapsi con altri neuroni, i neuroni nel cervello umano sono circa 86 miliardi e si stima che le sinapsi vadano da 100 mila miliardi a 1 milione di miliardi

- Celegans: un verme con 300 neuroni, il suo sistema nervoso è stato mappato completamente

## neurone di pitts e mcculloch
forniva input e putput binari, completamente digitali.

questa cosa è stata migliorata nel 1957 da frank rosenblatt che iventa il percettone, il quale ha un output, degli input e uan fuonzione matematica, output e input possono essere numeri reali, in questi percettroni ogni input ha un peso: ossia quanto quel collegamento è importante (+ è grande + è importante), è un numero

quando si parla dei parametri di chat gpt si parla dei pesi.

questo è importante perchè il nostro cervello impara e quindi si modifica, non si modificano i neuroni, si modifica nei collegamenti (si creano e si distruggono)

- le reti neruali artificiali apprendono cambiando il peso dei collegamenti

- se un input ha un peso =0 è come se non fosse collegato
