```python
emozioni=[]#da togliere questi numeri
immagini=[]

for i,val in enumerate(righe):
    riga=righe[i].split(",")
    if i!=0:
        emozioni.append(riga[0])
        immagini.append(np.array(riga[1].split(" "), dtype=np.int64).reshape(48,48))
```

```python
emozioni_complete=np.array(emozioni)
emozioni=np.unique(np.array(emozioni))
immagini=np.array(immagini)
```
