#### ESERCIZIO 2
Genera un array di numpy contenente i primi 100 numeri primi.

```python
lista_numP=[]
i=1
num0=0
while len(lista_numP) <101:
    for y in range(1,i+1):
        if i%y==0:
            num0=num0+1
    if num0==2:
        lista_numP.append(i)
    num0=0
    i=i+1
    
    
numPrimi=np.array(lista_numP)
print(numPrimi)
```

#### ESERCIZIO 4
Genera un array di numpy contenente i primi 10 numeri della successione di Fibonacci.

```python
lista_fibo=[]
while len(lista_fibo)<11:
    if len(lista_fibo)<2:
        lista_fibo.append(1)
    else: 
        lista_fibo.append(lista_fibo[len(lista_fibo)-1]+lista_fibo[len(lista_fibo)-2])
fibo=np.array(lista_fibo)
for f in fibo:
    print(f)
```

Usa numpy per creare dei dati simulati riferiti alla seguente situazione (in certi casi i dati simulati sono una buona approssimazione della realtà).
Nel parco vicino alla scuola è installata una centralina per la misurazione del livello di polveri sottili PM10 nell'aria. Talvolta la centralina va in uno stato di errore per cui la misurazione delle PM10 fallisce e il dato corrispondente riporta il valore 9999,9 all'interno dell'array delle misurazioni dati. Crea un array di 1000 campioni che generino dei valori secondo una distribuzione normale con valore medio 19 e varianza 5 e supponi che la centralina sia andata in errore 20 volte. La distribuzione gaussiana è una buona approssimazione dei dati reali della centralina? Motiva la risposta.

```python
dati=np.random.normal(19,5,1000)
valori_da_sostituire= np.random.choice(dati,20, replace=False)
for valore in valori_da_sostituire:
    indice = np.where(dati == valore)
    #print(dati[indice])
    dati[indice]=float(9999.9)
    #print(dati[indice])
    
for dato in dati:
    print(dato)
```
