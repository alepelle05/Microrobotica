Scarica MonthElb_012010_004028900.csv. In esso ci sono le misure meteorologiche dell'anno
2010 con cadenza oraria, per una stazione meteo italiana. Le colonne del file .csv sono le seguenti, da sinistra verso destra.
- 1 DATA;
- 2 ORA: espressa nel sistema UTC;
- 3 ID_RETE_MONIT: identificativo della Rete di Moni-toraggio;

- 6 VELV: velocità del vento;
- 7 DIRV: direzione del vento;
- 8 DEVST: deviazione standard della velocità trasver-sale;
- 9 BARO: pressione atmosferica;
- 10 TERMA: temperatura dell'aria;
- 11 IGRO: umidità relativa;
- 12 RADD: radiazione solare diretta;
- 13 RADR: radiazione solare riflessa;
- 14 PLUV: precipitazione.

In caso di valori mancanti è riportato il numero 9999.9.
Leggi il file csv come descritto in questo capitolo e crea e considera soltanto le seguenti grandezze presenti nel file: velocità del vento, pressione atmosferica, temperatura dell'aria, umidità relativa, precipitazione.
Considera queste 5 grandezze e crea una figura che rappresenti i 20 possibili diagrammi a dispersione e i 5 possibili istogrammi. La figura dovrà contenere 5 x 5 assi.

```python
dati=pd.read_csv("./MonthElb_012010_004028900.csv",header=None,names=["DATA","ORA",
                                                                      "ID","COMUNE",
                                                                      "NUM","VENTO_V",
                                                                     "VENTO_D","VENTO_DEV","PRESSIONE",
                                                                     "TEMP_ARIA","UMIDITA",
                                                                     "SOLARE_DIR","SOLARE_RIF",
                                                                      "PIOGGIA"])#tipo data frame
dati.head()#mi mostra la testa dei dati
```

```python
intestazione=list(dati.columns)
intestazione
```

```python
velocita_vento=dati["VENTO_V"].values
colonna_v=np.unique(velocita_vento)
velocita_vento
```

```python
pressione=dati["PRESSIONE"].values
colonna_p=np.unique(pressione)
pressione
```

```python
temperatura=dati["TEMP_ARIA"].values
colonna_t=np.unique(temperatura)
temperatura
```

```python
umidita=dati["UMIDITA"].values
colonna_u=np.unique(umidita)
umidita
```

```python
prec=dati["PIOGGIA"].values
colonna_prec=np.unique(prec)
prec
```

```python
label=["velocita_vento","pressione","temperatura","umidità","precipitazioni"]
```

```python
tem_div=temperatura!=9999.9
velocita_vento=velocita_vento[tem_div]
pressione=pressione[tem_div]
temperatura=temperatura[tem_div]
umidita=umidita[tem_div]
prec=prec[tem_div]
```

```python
grand=5
colonne_i=[colonna_v,colonna_p,colonna_t,colonna_u,colonna_prec]
fatures=[velocita_vento,pressione,temperatura,umidita,prec]

fig, axs=plt.subplots(nrows=grand, ncols=grand, figsize=(20,15))

for i,f1 in enumerate(fatures):
    for j,f2 in enumerate(fatures):
        if i==j:
            axs[i,j].hist(f1, bins=len(colonne_i[i]),color="k")
            axs[i,j].set_xlabel(label[i],fontsize=5)
            axs[i,j].set_ylabel("frequenza assoluta",fontsize=5)
        else:
            axs[i,j].scatter(f1,f2, s=1,c="r")
            axs[i,j].set_xlabel(label[i],fontsize=5)
            axs[i,j].set_ylabel(label[i],fontsize=5)
plt.show()
```

Utilizza numpy per creare una matrice 100 x 100 di numeri interi compresi tra 0 e 255. Crea una figura con due grafici, cioè due sottografici:
› il primo rappresenta la matrice con la mappa di colore che preferisci;
› il secondo rappresenta la distribuzione dei valori degli elementi della matrice come un istogramma

```python
matrice=np.random.randint(0,256,(100,100))
matrice
```

```python
fig, axs=plt.subplots(nrows=1, ncols=2, figsize=(10,3))



axs[0].imshow(matrice, cmap='Reds_r')
axs[0].set_title('Matrice con Mappa di Colore')
axs[0].axis("off")


axs[1].hist(matrice.flatten(), bins=np.size(np.unique(matrice.flatten())), color='blue')
axs[1].set_title('Distribuzione dei Valori')
axs[1].set_xlabel('Valore')
axs[1].set_ylabel('Frequenza')
plt.show()
```
